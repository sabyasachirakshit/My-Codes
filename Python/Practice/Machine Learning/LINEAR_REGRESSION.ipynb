```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv("E:\My Codes\Python\Practice\Machine Learning\Linear_Regression.csv")
df
```

```python
plt.plot(df[['Sales']],df[['Advertising']],color='r',marker='+')
plt.xlabel("Sales---->")
plt.ylabel("Advertising---->")
plt.title("Sales Advertisement Graph")
plt.show()
```

```python
model=LinearRegression()
model.fit(df[['Year','Sales']],df[['Advertising']])
score=model.score(df[['Year','Sales']],df[['Advertising']])*100
print(f"Model's Accuracy: {round(score,2)}%")
```
